```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_image(image_1,image_2,title_1='Original Image',title_2='New image'):
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,1)
    plt.imshow(image_1,cmap='gray')
    plt.title(title_1)
    plt.subplot(1,2,2)
    plt.imshow(image_2,cmap='gray')
    plt.title(title_2)
    plt.show()
```

```python
def plot_hist(old_image,new_image,title_old='Original image',title_new='New image'):
    intensity_val=np.array([x for x in range(256)])
    plt.subplot(1,2,1)
    plt.bar(intensity_val,cv2.calcHist([old_image],[0],None,[256],[0,256])[:,0],width=5)
    plt.title(title_old)
    plt.xlabel('Intensity')
    plt.subplot(1,2,2)
    plt.bar(intensity_val,cv2.calcHist([new_image],[0],None,[256],[0,256])[:,0],width=5)
    plt.title(title_new)
    plt.xlabel('Intensity')
    plt.show()
```

```python
def thresholding(input_img,threshold,max_value=255,min_value=0):
    N,M=input_img.shape
    image_out=np.zeros((N,M),dtype=np.uint8)
    
    for i in range(N):
        for j in range(N):
            if(input_img[i,j]>threshold):
                image_out[i,j]=max_value
            else:
                image_out[i,j]=min_value
        
    return image_out         
```

```python
cameraman=cv2.imread('camerman.jfif',cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(10,10))
plt.imshow(cameraman,cmap="gray")
plt.show()
```

```python
hist = cv2.calcHist([cameraman], [0], None, [256], [0, 256])
plt.bar(intensity_values, hist[:, 0], width=5)
plt.title("Bar histogram")
plt.show()
```

```python
threshold=87
max_value=255
min_value=0
new_image=thresholding(cameraman,threshold=threshold,max_value=max_value,min_value=min_value)
```

```python
plot_image(cameraman,new_image)
```

```python
ret,new_image=cv2.threshold(cameraman,threshold,max_value,cv2.THRESH_BINARY)
plot_image(cameraman,new_image)
plot_hist(cameraman,new_image)
```

```python
ret, otsu = cv2.threshold(cameraman,0,255,cv2.THRESH_OTSU)
plot_image(cameraman,otsu,"Orignal","Otsu")
plot_hist(cameraman, otsu,"Orignal"," Otsu's method")
```
